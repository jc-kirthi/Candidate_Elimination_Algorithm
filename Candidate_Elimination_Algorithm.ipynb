```python
import pandas as pd
data=pd.DataFrame([
    ["sunny","hot","high","weak","no"],
    ["sunny","hot","high","strong","no"],
    ["overcast","hot","high","weak","yes"],
    ["rain","mild","high","weak","yes"],
    ["rain","cool","normal","weak","yes"],
    ["rain","cool","normal","strong","no"],
    ["overcast","cool","normal","strong","yes"],
    ["sunny","mild","high","weak","yes"],
    ["sunny","cool","normal","weak","yes"],
    ["rain","mild","normal","weak","yes"],
    ["sunny","mild","normal","strong","yes"],
    ["overcast","mild","high","strong","yes"],
    ["overcast","hot","normal","weak","no"],
    ["rain","mild","high","strong","no"],
    ], columns=["outlook","temp","humidity","wind","playtennis"
    ])
print("traning data:\n")
print(data)
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
for i,label in enumerate(y):
  if label=='yes':
    s=x[i].copy()
    break
g=[["?" for _ in range(len(s))] for _ in range(len(s))]

print("\n initial specific hypothesis:",s)
print("initial genral hypothesis:",g)
for i,instance in enumerate(x):
   if y[i]=="yes":
      for j in range(len(s)):
         if instance [j]!=s[j]:
           s[j]="?"
   else:
      for j in range(len(s)):
         if instance [j]!=s[j]:
            g[j][j]=s[j]
         else:
            g[j][j]="?"

   print(f"\n step {i+1}-instance : {instance},label:{y[i]}")
   print("specific hypothesis:",s)
   print("genral hypothesis:",g)

g_final = [gh for gh in g if gh != ["?" for _ in range(len(s))]]
print("final specific hypothesis:",s)
print("final genral hypothesis:",g_final)
```
